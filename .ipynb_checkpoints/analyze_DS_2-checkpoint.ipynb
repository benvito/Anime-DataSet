```python
import pandas as pd
import seaborn as sns
import numpy as np
import openpyxl
import matplotlib.pyplot as plt
```

```python
exl = pd.read_excel ('Ds_Anime_2(500 anime).xlsx')
df = pd.DataFrame(exl)
df.head()
```

```python
sns.set_style("darkgrid")
rm_tgt_corr=round(df['year'].corr(df['average rating']), 3)
plt.figure(figsize=(9,6))
sns.lineplot(data=df, x="year", y="average rating", color='orange', alpha=0.4)
plt.scatter(x=df['year'], y=df['average rating'], alpha=0.6, s=80, color='red')
plt.title(f'Rating and Year (Correlation {rm_tgt_corr})', fontsize=14)
plt.xlabel('Rating')
plt.ylabel('Year')
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(9,6))
plt.plot([df['year'].min(), df['year'].max()], [df['average rating'].min(), df['average rating'].max()], color='orange')
plt.scatter(x=df['year'], y=df['average rating'], alpha=0.6, s=80, color='red')
plt.title(f'Rating and Year(linear regression)', fontsize=14)
plt.xlabel('Rating')
plt.ylabel('Year')
plt.show()
```

```python
sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
```
